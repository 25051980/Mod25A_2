## 1- Calcule de forma numérica a integral da função $x^3$ avaliada entre os pontos $1$ e $2$. Identifique o número de retângulos que você usou.  


Este código aproxima a integral da função \( f(x) = x^3 \) no intervalo \([1, 2]\) utilizando o método da soma dos retângulos. O intervalo é dividido em \( n = 50 \) retângulos, onde a área de cada retângulo é calculada como o produto da altura (\( f(x) \)) e a largura (\( dx \)). A soma dessas áreas fornece uma aproximação para a integral definida. No gráfico, a curva da função é representada em azul, enquanto os retângulos que aproximam a área sob a curva estão em laranja. Este método visualiza como os retângulos se aproximam da área real sob a curva.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definindo a função
def func(x):
    return x**3

# Configurando os parâmetros para o cálculo da integral
a = 1  # Limite inferior
b = 2  # Limite superior
n = 50  # Número de retângulos
dx = (b - a) / n  # Largura de cada retângulo

# Calculando a soma das áreas dos retângulos
x = np.linspace(a, b, n, endpoint=False)  # Posições dos retângulos (bordas esquerdas)
areas = func(x) * dx  # Altura vezes largura
integral = np.sum(areas)

# Exibindo o resultado
print("Número de retângulos:", n)
print("Valor aproximado da integral:", integral)

# Gerando o gráfico
x_curve = np.linspace(a, b, 500)  # Posições para desenhar a curva
y_curve = func(x_curve)

plt.figure(figsize=(8, 5))
plt.plot(x_curve, y_curve, label="f(x) = x³", color="blue")  # Curva da função
plt.bar(x, func(x), width=dx, align='edge', color='orange', edgecolor='black', alpha=0.6, label="Retângulos (área)")  # Retângulos
plt.title("Aproximação da Integral pela Soma de Retângulos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```

A imagem mostra a aproximação da integral da função \( f(x) = x^3 \) no intervalo \([1, 2]\) utilizando o método da soma dos retângulos. Com \( n = 50 \) retângulos, foi obtido um valor aproximado para a integral de \( 3.6803 \). Os retângulos representam as áreas calculadas em cada subdivisão do intervalo, enquanto a curva azul ilustra a função original. O método demonstra como a soma das áreas dos retângulos converge para o valor da integral definida conforme o número de retângulos aumenta.


## 2- Calcule essa área de forma analítica.

A área sob a curva da função \( f(x) = x^3 \) no intervalo \([1, 2]\) foi calculada de forma analítica utilizando a integral definida. A integral de \(x^3\) é dada por \( \int x^3 dx = \frac{x^4}{4} + C \). Avaliando a integral no intervalo \([1, 2]\), obtemos:

\[
\text{Área} = \left[\frac{x^4}{4}\right]_1^2 = \frac{2^4}{4} - \frac{1^4}{4} = \frac{16}{4} - \frac{1}{4} = \frac{15}{4} = 3.75
\]

Assim, a área sob a curva entre os pontos 1 e 2 é exatamente **3.75**.


```python
# Cálculo da área de forma analítica
a = 1  # Limite inferior
b = 2  # Limite superior

# Calculando a integral analítica
area_analitica = (b**4 / 4) - (a**4 / 4)

# Exibindo o resultado
print("Área calculada de forma analítica:", area_analitica)
```
